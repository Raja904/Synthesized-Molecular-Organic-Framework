### We are going to use SVM

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df1.head()
```

```python
x1 = df1.drop(columns='NCH4 0.1 Bar (mol/kg)')
y1 = df1['NCH4 0.1 Bar (mol/kg)']
```

```python
x1.shape
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.33, random_state=42)
```

```python
X1_train.shape
```

```python
y1_test.shape
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

# Define the parameter grid for GridSearchCV
param_grid = {
    'svr__kernel': ['linear', 'rbf'],
    'svr__C': [0.1, 1, 10],
    'svr__gamma': ['scale', 'auto'],
    'svr__epsilon': [0.1, 0.2, 0.5]
}

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the model to the training data
grid_search.fit(X1_train, y1_train)

# Print the best parameters and the corresponding model's score
print("Best parameters found: ", grid_search.best_params_)
print("Best model score: ", grid_search.best_score_)
```

```python
pipeline1 = Pipeline(steps=[('standardscaler', StandardScaler()),
                ('svr', SVR(C= 10, epsilon= 0.1, gamma= 'scale', kernel= 'linear'))])
```

```python
pipeline1.fit(X1_train,y1_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming you have already trained your model and made predictions
predictions1 = pipeline1.predict(X1_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y1_test, predictions1)
print(f'Mean Squared Error: {mse:.4f}')

# Calculate Root Mean Squared Error
rmse = mean_squared_error(y1_test, predictions1, squared=False)
print(f'Root Mean Squared Error: {rmse:.4f}')

# Calculate Mean Absolute Error
mae = mean_absolute_error(y1_test, predictions1)
print(f'Mean Absolute Error: {mae:.4f}')

# Calculate R-squared
r2 = r2_score(y1_test, predictions1)
print(f'R-squared: {r2:.4f}')

```

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming you have predictions1 to predictions6 and corresponding y_test values
predictions_list = [predictions1, predictions2, predictions3, predictions4, predictions5, predictions6]
y_test_list = [y1_test, y2_test, y3_test, y4_test, y5_test, y6_test]

# Define gas and pressure conditions
conditions = ['CH4 at 0.1bar', 'CH4 at 1bar', 'CH4 at 10bar', 'N2 at 0.1bar', 'N2 at 1bar', 'N2 at 10bar']

# Create a DataFrame to store the error metrics
error_metrics_df = pd.DataFrame(columns=['Model', 'Gas and Pressure Condition', 'Mean Squared Error', 'Root Mean Squared Error', 'Mean Absolute Error', 'R-squared'])

# Loop through the predictions and calculate metrics
for i, (predictions, y_test) in enumerate(zip(predictions_list, y_test_list)):
    mse = mean_squared_error(y_test, predictions)
    rmse = mean_squared_error(y_test, predictions, squared=False)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)

    model_name = 'SVR'  # Assuming the model type is SVR

    # Append the metrics to the DataFrame
    error_metrics_df = error_metrics_df.append({
        'Model': model_name,
        'Gas and Pressure Condition': conditions[i],
        'Mean Squared Error': mse,
        'Root Mean Squared Error': rmse,
        'Mean Absolute Error': mae,
        'R-squared': r2
    }, ignore_index=True)

# Display the error metrics DataFrame
markdown_table = error_metrics_df.to_markdown(index=False)
print(markdown_table)
```
